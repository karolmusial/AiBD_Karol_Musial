## 7. <b>Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.<b>

```python
# ID Polski Nigerii i Bangladeszu
df = pd.read_sql("SELECT * FROM country WHERE country in ('Poland', 'Nigeria', 'Bangladesh')",con=connection)
country_ids = tuple(df['country_id'].tolist())

# ID wszystkich miast z powyższych państw
df = pd.read_sql("SELECT * FROM city WHERE country_id in {}".format(country_ids),con=connection)
city_ids = tuple(df['city_id'].tolist())

# ID wszystkich adresów z tych miast
df = pd.read_sql("SELECT * FROM address WHERE city_id in {}".format(city_ids),con=connection)
address_ids = tuple(df['address_id'].tolist())

# Wszyscy klienci o danych adress_id
df = pd.read_sql("SELECT * FROM customer WHERE address_id in {}".format(address_ids), con=connection)
df
```

### <b>Zadanie rozpoczęto od znalezienia 'country_id' Polski Nigerii i Bangladeszu, zapisano je do listy. Następnie znaleziono 'city_id' wszystkich miast w tych krajach, to samo zrobiono z 'adress_id' adresów. Ostatnim etapem było wypisanie tablicy wszystkich klientów których 'adress_id' znajduje się w liście adresów.<b>

## 10. <b>Jakie kategorie filmów zostały wypożyczone przez klientów?<b>

```python
# Wybrano wszystkie wypożyczone kiedykolwiek inventory_id
df = pd.read_sql("SELECT DISTINCT inventory_id FROM rental ",con=connection)
inventory_ids = tuple(df['inventory_id'].tolist())

# film_id odpowiadające znalezionym inventory_id
df = pd.read_sql("SELECT * FROM inventory WHERE inventory_id in {}".format(inventory_ids),con=connection)
film_ids = tuple(df['film_id'].tolist())

# category_id odpowiadające znalezionym film_id
df = pd.read_sql("SELECT * FROM film_category WHERE film_id in {}".format(film_ids),con=connection)
category_ids = tuple(df['category_id'].tolist())

# Wszystkie kategoriie o danych category_id
df = pd.read_sql("SELECT * FROM category WHERE category_id in {}".format(category_ids),con=connection)
df
```

### <b>By odnaleść wszystkie wypożyczone kategorie filmów, rozpoczęto od znalezienia wszystkich 'inventory_id', następnie przypisano odpowiadające im 'film_id', tym z kolei 'category_id'. W ten sposób możliwe było wypisanie wszystkich kategorii które znajdowały się na liście. Okazało się (jak można się było spodziewać), że wszystkie kategorie były wypożyczone.<b>

## 11. <b>Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.<b>

```python
# country_id USA
df = pd.read_sql("SELECT * FROM country WHERE country = 'United States'",con=connection)
country_id = tuple(df['country_id'].tolist())

# city_id wszystkich miast z USA
df = pd.read_sql("SELECT * FROM city WHERE country_id = {}".format(country_id[0]),con=connection)
city_ids = tuple(df['city_id'].tolist())

# adress_id wszystkich adresów z USA
df = pd.read_sql("SELECT * FROM address WHERE city_id IN {}".format(city_ids),con=connection)
address_ids = tuple(df['city_id'].tolist())

# customer_id wszystkich klientów z USA
df = pd.read_sql("SELECT * FROM customer WHERE address_id IN {}".format(address_ids),con=connection)
customer_ids = tuple(df['customer_id'].tolist())

# inventory_id wszystkich wypożyczeń klientów z USA
df = pd.read_sql("SELECT * FROM rental WHERE customer_id IN {}".format(customer_ids),con=connection)
invetory_ids = tuple(df['inventory_id'].tolist())

# film_id wszystkich inventory z poprzedniej listy
df = pd.read_sql("SELECT * FROM inventory WHERE inventory_id in {}".format(inventory_ids),con=connection)
film_ids = tuple(df['film_id'].tolist())

# category_id wszystkich kategorii o danych filmach
df = pd.read_sql("SELECT * FROM film_category WHERE film_id in {}".format(film_ids),con=connection)
category_ids = tuple(df['category_id'].tolist())

# Tablica kategorii z listy
df = pd.read_sql("SELECT * FROM category WHERE category_id in {}".format(category_ids),con=connection)
df
```

### <b>Wykonanie polecenia podzielono na 8 etapów. Znalezienie 'country_id' USA, znalezienie 'city_id' wszystkich miast z USA, znalezienie 'adress_id' wszystkich adresów z USA oraz znalezienie 'customer_id' wszystkich klientów o tych adresach. W ten sposób otrzymano listę wszystkich klientów mieszkających w USA. W kolejnych etapach zebrano 'inventory_id' wszystkich wypożyczeń tych klientów, odpowiadające im 'film_id', 'category_id' i ostatecznie wypisano wszystkie kategorie.<b>

## 12. <b>Znajdź wszystkie tytuły filmów, w których grał: Olympia Pfeiffer lub Julia Zellweger lub Ellen Presley<b>

```python
# actor_id podanych aktorów
df = pd.read_sql("SELECT * FROM actor WHERE first_name ='Olympia' AND last_name ='Pfeiffer' OR first_name ='Julia' AND last_name ='Zellweger' OR first_name ='Ellen' AND last_name ='Presley' ",con=connection)
actor_ids = tuple(df['actor_id'].tolist())

# film_id filmów które zawierają danych aktorów
df = pd.read_sql("SELECT * FROM film_actor WHERE actor_id in {}".format(actor_ids),con=connection)
film_ids = tuple(df['film_id'].tolist())

# Wypisanie wszystkich filmów
df = pd.read_sql("SELECT title FROM film WHERE film_id in {}".format(film_ids),con=connection)
df
```

### <b>Korzystając z zagnieżdżonej konstrukcji 'WHERE' oraz operatorów logicznych, wydobyto 'actor_id' podanych w poleceniu aktorów, przypisano je do listy filmów i otrzymano listę filmów w których oni grali.<b>
