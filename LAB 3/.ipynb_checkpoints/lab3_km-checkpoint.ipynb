```python
# 10. Jakie kategorie zostaly wypozyczone
df = pd.read_sql("SELECT DISTINCT inventory_id FROM rental ",con=connection)
inventories_id = tuple(df['inventory_id'].tolist())

df = pd.read_sql("SELECT * FROM inventory WHERE inventory_id in {}".format(inventories_id),con=connection)
films_id = tuple(df['film_id'].tolist())

df = pd.read_sql("SELECT * FROM film_category WHERE film_id in {}".format(films_id),con=connection)
categories_id = tuple(df['category_id'].tolist())

df = pd.read_sql("SELECT * FROM category WHERE category_id in {}".format(categories_id),con=connection)
df
```

```python
#11. Wszystkie kategorie ze stanow
df = pd.read_sql("SELECT * FROM country WHERE country = 'United States'",con=connection)
country_id = tuple(df['country_id'].tolist())

df = pd.read_sql("SELECT * FROM city WHERE country_id = {}".format(country_id[0]),con=connection)
cities_id = tuple(df['city_id'].tolist())

df = pd.read_sql("SELECT * FROM address WHERE city_id IN {}".format(cities_id),con=connection)
addresses_id = tuple(df['city_id'].tolist())

df = pd.read_sql("SELECT * FROM customer WHERE address_id IN {}".format(addresses_id),con=connection)
customers_id = tuple(df['customer_id'].tolist())

df = pd.read_sql("SELECT * FROM rental WHERE customer_id IN {}".format(customers_id),con=connection)
invetories_id = tuple(df['inventory_id'].tolist())

df = pd.read_sql("SELECT * FROM inventory WHERE inventory_id in {}".format(inventories_id),con=connection)
films_id = tuple(df['film_id'].tolist())

df = pd.read_sql("SELECT * FROM film_category WHERE film_id in {}".format(films_id),con=connection)
categories_id = tuple(df['category_id'].tolist())

df = pd.read_sql("SELECT * FROM category WHERE category_id in {}".format(categories_id),con=connection)
df
```
