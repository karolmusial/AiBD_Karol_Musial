```python
# Importowanie
from matplotlib import pyplot as plt
import pandas as pd

# Wczytywanie danych
data = pd.read_csv('../Original Data/originaldata.csv')
DF = pd.DataFrame(data)
```

```python
# Wstępna obróbka
DF.columns = ['Index', 'Days', 'Brand', 'Age',
       'Sex', 'Score']
DF = DF.drop(['Index'], axis=1)
DF
```

```python
# Płeć kupującego
DF["Sex"] = DF['Sex'].map({'bd.' : None, 'M' : 'M', 'K' : 'K'})
DF
```

```python
# Zapisywanie zmodyfikowanych danych do pliku "new_data.csv"
DF.to_csv('../Analysis Data/new_data.csv', index=False)
```
