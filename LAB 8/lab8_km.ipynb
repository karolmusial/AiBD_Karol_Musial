```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_absolute_error, mean_squared_error, classification_report
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

db_string = "postgresql://wbauer_adb:adb2020@pgsql-196447.vipserv.org:5432/wbauer_classifier_abd"
db = create_engine(db_string)
```

```python
# Popularnosc kategorii
select_string="SELECT setseed(0.5); SELECT genres.name, count(movies) FROM movies JOIN genres_movies "\
              "ON movies.movie_id = genres_movies.movie_id JOIN genres ON genres_movies.genre_id = genres.genre_id "\
              "WHERE random() > 0.9 GROUP BY genres.name ORDER BY count(movies) DESC"  
df=pd.read_sql(select_string,con=db)
df.head(10)

# Wybrano kategorie numer 1 - Drama
```

```python
# Baza danych filmów
select_string="SELECT setseed(0.5); SELECT movies.movie_name, movies.plot, genres.name FROM movies JOIN genres_movies "\
              "ON movies.movie_id = genres_movies.movie_id JOIN genres ON genres_movies.genre_id = genres.genre_id "\
              "WHERE random() > 0.9"  

df=pd.read_sql(select_string,con=db)
df
```

```python
# Kolumna przypisujaca 1 lub 0 w zaleznosci od gatunku
df['drama'] = 0
df.loc[df.name == 'Drama', 'drama'] = 1
df
```

```python
# Tworzenie bag of words
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(lowercase=True, strip_accents='ascii', stop_words='english', max_features=20)

# Przekształcanie opisów do zadanej postaci
X = vectorizer.fit_transform(df['plot'])
y = df['drama']

# Dzielenie na zbiory uczące i testujące
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Tworzenie modelu
LRM = LogisticRegression(max_iter = 1000)
LRM.fit(X_train, y_train)

# Wypisywanie wskaznikow
print("\nCoef = {a}".format(a=LRM.coef_))
print("Intercept = {a}\n".format(a=LRM.intercept_))

# Predykcja czesci testowej
y_pred = LRM.predict(X_test)
print("Roznica miedzy predykcja a rzeczywistym wynikiem : \n{a}\n".format(a=(y_pred - y_test)))

# Walidacja danych
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
cr = classification_report(y_test, y_pred)
print('\nŚredni błąd bezwzględny = {a}'.format(a=mae))
print('Błąd średniokwadratowy = {a}'.format(a=mse))
print('Pierwiastek błędu średniokwadratowego = {a}\n'.format(a=rmse))
print('Raport:\n{a}\n\n'.format(a=cr))

# Confusion Matrix
c_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(c_matrix)
```

```python
# Prawdopodobienstwa 
predict_probabilities = LRM.predict_proba(X)[:,1]
fpr, tpr, _ = roc_curve(y, predict_probabilities)
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Krzywa ROC')
plt.show()
```
