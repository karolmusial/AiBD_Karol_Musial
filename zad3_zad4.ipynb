<center><h2>Zadanie 3</h2></center>

Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

$x>-1 ~oraz~ x<1$

$x>-6 ~oraz~ x<6$

$x>0 ~oraz~ x<5$

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

Wskazówki: Deklaracja funkcji w Python.

W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu Matplotlib.

W pliku mają być zawarte: imię, nazwisko, opis wykonywanego ćwiczenia - formatowanie markdown w poszczególnych komórkach.

```python
# Importowanie bibliotek
import numpy as np
from matplotlib import pyplot as plt

# Definiowanie funkcji
def f(x):
    return x**2 + 5

# Tworzenie przedziałów
x1 = np.linspace(-1, 1, 1000)
x2 = np.linspace(-6, 6, 1000)
x3 = np.linspace(0, 5, 1000)

# Rysowanie wykresu dla każdego przedziału
plt.plot(x1, f(x1))
plt.title('Funkcja w przedziale x>-1 oraz x<1')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.legend(['funkcja f'])
plt.show()

plt.plot(x2, f(x2))
plt.title('Funkcja w przedziale x>-6 oraz x<6')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.legend(['funkcja f'])
plt.show()

plt.plot(x3, f(x3))
plt.title('Funkcja w przedziale x>0 oraz x<5')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.legend(['funkcja f'])
plt.show()
```

Utwórz dataframe, w którym kolumny mają nazwy: name, surname, age, sex. Uzupełnij pięcioma dowolnymi rekordami oraz wyświetl informacje o danych pandas inf, opis danych pandas describe, wyświetl pierwsze trzy rekordy pandas head.

```python
# Import potrzebnej biblioteki
import pandas as pd

# Tworzenie dataframe
names = ['Feanor', 'Fingolfin', 'Turgon', 'Fingon', 'Finrod']
surnames = ['-', '-', '-', '-', '-']
data = {'name':names, 'surname':surnames, 'age':['3142', '3426', '2426', '2771', '2381'], 'sex':['MALE', 'MALE', 'MALE', 'MALE', 'MALE']}
df = pd.DataFrame(data)

# Wyświetlanie Data Frame
print(df)
```

```python
# Wyświetlanie informacji o danych
df.info(verbose=True)
```

```python
# Wyświetlanie opisu danych
df.describe()
```

```python
# Wyświetlanie pierwszych trzech rekordów
df.head(3)
```
